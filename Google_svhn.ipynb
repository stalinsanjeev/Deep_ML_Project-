```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
import scipy.io

def load_svhn_data(path):
    data = scipy.io.loadmat(path)
    images = data['X']
    labels = data['y']
    images = images.transpose((3,0,1,2))
    labels[labels == 10] = 0
    return images, labels


train_images, train_labels = load_svhn_data('/content/drive/My Drive/train_32x32.mat')
test_images, test_labels = load_svhn_data('/content/drive/My Drive/test_32x32.mat')


train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255


train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)


model = Sequential([
    Flatten(input_shape=(32, 32, 3)),
    Dense(128, activation='relu'),
    Dropout(0.0),
    Dense(128, activation='relu'),
    Dropout(0.0),
    Dense(10, activation='softmax')
])


model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])


history1 = model.fit(train_images, train_labels, epochs=20, batch_size=32, validation_data=(test_images, test_labels))


test_loss, test_acc = model.evaluate(test_images, test_labels, )
print('Test accuracy:', test_acc)


train_acc = history1.history['accuracy']
test_acc = history1.history['val_accuracy']
epochs = range(1, len(train_acc) + 1)


plt.plot(epochs, train_acc, 'bo-', label='Training accuracy')
plt.plot(epochs, test_acc, 'ro-', label='Test accuracy')
plt.title('Training and Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
